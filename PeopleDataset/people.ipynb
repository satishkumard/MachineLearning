Q1) Load the dataset into a pandas dataframe and display the first 5 lines of the dataset along with the column headings. Note that the data does not come along with the column headings, so you should be adding that to the data frame.


```python
import pandas as pd
import numpy as np
import os

#Loading the people file
peopleDF = pd.read_csv('people.csv', ',', 
                  names = ["FirstName", "LastName", "Age", "CompanyName", "StreetName", "City", "County", "State", "PhoneNumber", "ZipCode", "Email", "URL"])

#Display the first 5 lines of the dataset
peopleDF.head()
```

Q2) Drop the county column.

```python
#Drop the county column
peopleDF.drop(['County'], axis=1)
```

Q3) Keep only those rows that have a minimum of 4 values, otherwise delete them.

```python
#Keep only those rows that have a minimum of 4 values, otherwise delete them.
peopleDF.dropna(thresh=4)
```

Q4) Delete rows with email missing.

```python
#Delete rows with email missing
peopleDF.dropna(subset=['Email'])
```

Q5) Impute the missing values in age with the mean of the column.

```python
#Impute missing values for Age with mean of the column
from sklearn.impute import SimpleImputer
X=peopleDF.iloc[:,:].values
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer.fit(X[:,[2]])
X[:,[2]] =imputer.transform(X[:,[2]])

print(X[:,[2]])  # shows new age column standalone
# print(X)       # shows new age column inline with other values
```
